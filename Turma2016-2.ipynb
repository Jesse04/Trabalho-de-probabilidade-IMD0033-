```python
#Abaixo importamos as bibliotecas que iremos usar para a construçao dos dataframes e os gráficos
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
#O primeiro dataframe que criamos refere-se aos cursos ofertados pela UFRN, a partir desta tabela
#encontramos o "id_curso" que corresponde ao código utilizado para identificar o curso.

cursos_url = 'http://dados.ufrn.br/dataset/02526b96-cf40-4507-90b0-3afe5ddd53e7/resource/a10bc434-9a2d-491a-ae8c-41cf643c35bc/download/cursos-de-graduacao.csv'
cursos = pd.read_csv(cursos_url,sep = ';', usecols=['id_curso', 'nome'])
cursos
```

```python
#Utilizamos o algoritmo abaixo para encontrarmos no dataframe "CURSOS" o código referente ao curso
#Tecnologia da Informação que é o curso que iremos analizar em nossa base de dados.
#Assim encontramos que a id_curso que iremos trabalhar é o número 92127264.

cursos.id_curso = cursos.id_curso.astype(int)
curso_id = int
for i,j in enumerate(cursos['nome']):
    if j == 'TECNOLOGIA DA INFORMAÇÃO':
        curso_id = cursos['id_curso'].iloc[i]
curso_id
```

```python
#O próximo dataframe que iremos criar é o referente as matrículas em componentes, pois nela temos dados
#referentes a turma, curso, discente e situação do discente (aprovado, reprovado, aprovado por nota).

matriculas_url = 'http://dados.ufrn.br/dataset/c8650d55-3c5a-4787-a126-d28a4ef902a6/resource/f6179838-b619-4d7d-af9c-18c438b80dd4/download/matriculas-de-2016.2.csv'
matriculas = pd.read_csv(matriculas_url, usecols=["id_turma","id_curso","descricao", "discente"], sep = ';')
matriculas
```

```python
#Como já sabemos qual a "id_curso" que queremos utilizar, filtramos o dataframe para que mostrassem apenas
#as turmas referentes ao curso de Tecnologia da Informação
matriculas = matriculas[matriculas.id_curso == 92127264]
matriculas
```

```python
#Em seguida criamos o dataframe "turmas" para que com ele conseguíssemos a situação referente
#a cada turma (se está consolidada, aberta).
turmas_url = 'http://dados.ufrn.br/dataset/1938623d-fb07-41a4-a55a-1691f7c3b8b5/resource/5e8e3228-7f22-40a2-9efd-561c44844567/download/turmas-2016.2.csv'
turmas = pd.read_csv(turmas_url, sep=';')
turmas
```

```python
#Neste dataframe buscamos os componentes curriculares presenciais, nele encontramos os nomes das disciplinas
#e a unidade responsável pela disciplina
componentes_url = 'http://dados.ufrn.br/dataset/3fea67e8-6916-4ed0-aaa6-9a8ca06a9bdc/resource/9a3521d2-4bc5-4fda-93f0-f701c8a20727/download/componentes-curriculares-presenciais.csv'
componentes = pd.read_csv(componentes_url, sep=';', usecols=["id_componente","codigo","nivel","nome","unidade_responsavel"])
componentes
```

```python
matriculas.info()
```

```python
#No dataframe "turmas" percebemos que a coluna id_componoente_curricular e a coluna id_componente do
#dataframe "componentes" se referem aos mesmos dados, por isso, fizemos a mudança no nome da coluna
#Para que pudéssemos realizar a coleta dos dados utilizamos a função merge para unir os dados do dataframe
#turmas e componentes, e denominamos o dataframe seguinte de turmas_imd
turmas.rename(columns={'id_componente_curricular': 'id_componente'}, inplace=True)
turmas.id_componente = turmas.id_componente.astype(str)
componentes.dropna(inplace=True)
componentes = componentes[~componentes.id_componente.duplicated()]
colunas = ["id_turma","id_docente_interno","codigo","nivel_ensino","nome","unidade_responsavel",
          "situacao_turma","capacidade_aluno","descricao_horario"]
merge_turmas_disciplinas = pd.merge(turmas, componentes, on='id_componente')[colunas]
merge_turmas_disciplinas.dropna()
```

```python
merge_turmas_disciplinas.info()
```

```python
#Neste dataframe realizamos mais uma união de dataframes
#Neste caso utilizamos o dataframe criado acima com o dataframe matriculas
columns = ['id_turma', 'id_docente_interno', 'codigo', 'nome', 'unidade_responsavel']
turmas_imd = pd.merge(matriculas, merge_turmas_disciplinas, on='id_turma')
#turmas_imd = turmas_imd[~turmas_imd.id_turma.duplicated()]
#turmas_imd
```
